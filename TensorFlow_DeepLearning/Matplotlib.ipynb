```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,20)
x
```

```python
y = x ** 2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'red')
```

```python
plt.plot(x,y,'g')
```

```python
plt.plot(x,y,'r*')
```

```python
plt.plot(x,y,'b--')
plt.title("T√≠tulo del Gr√°fico")
plt.xlabel("eje de los valores x")
plt.ylabel("eje y")
```
